```python
y = np.linspace(0,3, 2) #ignore this points for now
y
```

```python
y[0], y[1] = 0, 0.8
```

```python
y
```

```python
# How many points do you want to learn between 2 points?
i_want = 50 #points
```

```python
point = i_want+2
```

```python
a = np.linspace(0,0, i_want)
b = np.linspace(0,0, i_want)
```

```python
for i in range(i_want):
  x_step = (x[1]-x[0])/(point)
  a[i] = x[0] + (i+1)*x_step

  y_step = (y[1]-y[0])/(point)
  b[i] = y[0] + (i+1)*y_step

print("X Points")
for j in range(len(a)):
  print(a[j], "/n")
print ("Y Points")
for l in range(len(b)):
  print(b[l], "/n")
```
