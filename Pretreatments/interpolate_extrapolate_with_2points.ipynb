```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
x = np.array([438, 873.839])
y = np.array([-4.734305785, -7.864])
f = interpolate.interp1d(x, y, fill_value='extrapolate')
```

```python
xnew = np.arange(436, 886, 1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

```python
dat1={'x':xnew, 'z':ynew}
tail1=pd.DataFrame(data=dat1)
tail1.to_excel("tail00.xlsx")
tail1
```
