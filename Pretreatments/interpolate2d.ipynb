```python
from matplotlib.tri import Triangulation, CubicTriInterpolator
import rasterio
```

```python
triFn = Triangulation(x, y)
lintriFn = CubicTriInterpolator(triFn, z)
```

```python
step = 2
xcoor = np.arange(x.min(),x.max()+step, step)
ycoor = np.arange(y.min(),y.max()+step, step)
zcoor = np.zeros([ycoor.shape[0],xcoor.shape[0]])
```

```python
for indexX, x in np.ndenumerate(xcoor):
  for indexY, y in np.ndenumerate(ycoor):
    f = lintriFn(x,y)
    if f == f:
      zcoor[indexY,indexX] = f
    else:
      zcoor[indexY,indexX] = np.nan
plt.imshow(zcoor)
```
