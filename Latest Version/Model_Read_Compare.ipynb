```python
space = ["Hs"]*17 # Just for concat func.
df_space = pd.DataFrame({'Hs': space}).transpose() # Just for concat func.
```

```python
models = os.listdir()
for m in models: #for loop for entering all model folders
    os.chdir(os.path.join(main_dir,m))
    excels = glob.glob("*xlsx")
    df = pd.read_excel(excels[0]) #excels[0] is Default.xlsx excels[1] is Default+TRIAD.xlsx (Doesn't matter for measurement values but important for SWAN values.)
    measurement_values = ["Measurements",excels[0][0:8]]
    for t in range(len(df.iloc[59:60,3:19].values.flatten())): #for loop for debugging 'array in array' issue
        measurement_values.append(df.iloc[59:60,3:19].reset_index(drop=True).values.flatten().astype('float64')[t])
    df_space = pd.concat([df_space, pd.DataFrame({'Measurements': measurement_values}).transpose()], axis=0)
    os.chdir(main_dir)
```

```python
df_space = df_space.reset_index(drop=True)
df_meas = df_space.dropna(thresh=17) # If there are any missed values, remove those rows.
df_meas= df_meas.iloc[1:, :] # Remove unneccessary first row
df_meas # Measurement DF to print to excel file
```

```python
space = ["Hs"]*17 # Just for concat func.
df_space2 = pd.DataFrame({'Hs': space}).transpose() # Just for concat func.
models = os.listdir()
for m2 in models: #for loop for entering all model folders
    os.chdir(os.path.join(main_dir,m2))
    excels = glob.glob("*xlsx")
    df = pd.read_excel(excels[0]) #excels[0] is Default.xlsx excels[1] is Default+TRIAD.xlsx !!!! IMPORTANT
    model_values = ["SWAN", excels[0][0:8] + ' ' + model_name]
    for t2 in range(len(df.iloc[5:6,3:19].values.flatten())): # for loop for debugging 'array in array' issue
        model_values.append(df.iloc[5:6,3:19].reset_index(drop=True).values.flatten().astype('float64')[t2])
    df_space2 = pd.concat([df_space2, pd.DataFrame({'SWAN': model_values}).transpose()], axis=0)
    os.chdir(main_dir)
model_values = [excels[0][0:7] + ' ' + excels[0][9:12]]
df_space2 = df_space2.reset_index(drop=True)
df_space2 = df_space2.iloc[1:, :] # Remove unneccessary first row
drop_idx = df_space2[~df_space2.index.isin(df_meas.index)].index # Drop the values didn't include DF of measurements
df_SWAN = df_space2.drop(drop_idx) # Drop the values didn't include DF of measurements
df_SWAN # Measurement DF to print to excel file
```
