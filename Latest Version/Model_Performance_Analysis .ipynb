# Performance Analysis of SWAN BKD Models

This file will analyze SWAN BKD Models which made in 18-01-2023 (folder \models3)

```python
import pandas as pd
import os
import glob
import numpy as np
import statistics
import scipy.stats
import math
from sklearn.metrics import mean_squared_error
```

```python
main_dir = r"C:\Users\pc\OneDrive_GTU\OneDrive - Gebze Teknik Üniversitesi\Tez\Proje\models3"
os.chdir(main_dir)
```

```python
# MODELLERİN İSİMLERİNİ TANIMLA
# model_name = "Default BKD" # TRIAD modellerde excels[0]'ı değiştirmeyi unutma
```

```python
def Root_Mean(meas, pred):
    MSE = mean_squared_error(meas,pred)
    RMSE = math.sqrt(MSE)
    return MSE,RMSE
```

```python
space = ["Hs"]*17 # Just for concat func.
df_space = pd.DataFrame({'Hs': space}).transpose() # Just for concat func.
```

```python
models = os.listdir()
for m in models: #for loop for entering all model folders
    os.chdir(os.path.join(main_dir,m))
    excels = glob.glob("*xlsx")
    df = pd.read_excel(excels[0]) #excels[0] is Default.xlsx excels[1] is Default+TRIAD.xlsx (Doesn't matter for measurement values but important for SWAN values.)
    measurement_values = ["Measurements",excels[0][0:8], m[15:]]
    for t in range(len(df.iloc[59:60,4:19].values.flatten())): #for loop for debugging 'array in array' issue
        measurement_values.append(df.iloc[59:60,4:19].reset_index(drop=True).values.flatten().astype('float64')[t])
    df_space = pd.concat([df_space, pd.DataFrame({'Measurements': measurement_values}).transpose()], axis=0)
    os.chdir(main_dir)
```

```python
df_space = df_space.reset_index(drop=True)
df_meas = df_space.dropna(thresh=17) # If there are any missed values, remove those rows.
df_meas= df_meas.iloc[1:, :] # Remove unneccessary first row
df_meas = df_meas.sort_values(by=[2, 1]) # Sort models
df_meas # Measurement DF to print to excel file
```

```python
space = ["Hs"]*17 # Just for concat func.
df_space2 = pd.DataFrame({'Hs': space}).transpose() # Just for concat func.
models = os.listdir()
for m2 in models: #for loop for entering all model folders
    os.chdir(os.path.join(main_dir,m2))
    excels = glob.glob("*xlsx")
    df = pd.read_excel(excels[0]) #excels[0] is Default.xlsx excels[1] is Default+TRIAD.xlsx !!!! IMPORTANT
    model_values = ["SWAN", excels[0][0:8], m2[15:]]
    for t2 in range(len(df.iloc[5:6,4:19].values.flatten())): # for loop for debugging 'array in array' issue
        model_values.append(df.iloc[5:6,4:19].reset_index(drop=True).values.flatten().astype('float64')[t2])
    df_space2 = pd.concat([df_space2, pd.DataFrame({'SWAN': model_values}).transpose()], axis=0)
    os.chdir(main_dir)
model_values = [excels[0][0:7] + ' ' + excels[0][9:12]]
df_space2 = df_space2.reset_index(drop=True)
df_space2 = df_space2.iloc[1:, :] # Remove unneccessary first row
drop_idx = df_space2[~df_space2.index.isin(df_meas.index)].index # Drop the values didn't include DF of measurements
df_SWAN = df_space2.drop(drop_idx) # Drop the values didn't include DF of measurements
df_SWAN = df_SWAN.sort_values([2, 1])
df_SWAN # Measurement DF to print to excel file
```

```python
dif_model_types = df_SWAN[2].unique() # define all unique model types (parameters)
```

```python
names_of_models = []
Loc1_corr = []
Loc2_corr = []
Loc14_corr = []
Loc15_corr = []
MSE1 = []
RMSE1 = []
MSE2 = []
RMSE2 = []
MSE14 = []
RMSE14 = []
MSE15 = []
RMSE15 = []
STDEV1,STDEV2,STDEV14,STDEV15 = [],[],[],[]
STDEVm1,STDEVm2,STDEVm14,STDEVm15 = [],[],[],[]
bias1, bias2, bias14, bias15 = [],[],[],[]
var1, var2, var14, var15 = [],[],[],[]
for i in range(len(dif_model_types)):
    MEAS = df_meas.loc[df_meas[2] == dif_model_types[i]]
    SWAN = df_SWAN.loc[df_SWAN[2] == dif_model_types[i]]
    #print(MEAS[3].astype('float64'), SWAN[3].astype('float64'))
    names_of_models.append(MEAS[2].reset_index(drop=True)[0])
    Loc1_corr.append(np.corrcoef(MEAS[3].astype('float64'), SWAN[3].astype('float64'))[0][1])
    Loc2_corr.append(np.corrcoef(MEAS[4].astype('float64'), SWAN[4].astype('float64'))[0][1])
    Loc14_corr.append(np.corrcoef(MEAS[16].astype('float64'), SWAN[16].astype('float64'))[0][1])
    Loc15_corr.append(np.corrcoef(MEAS[17].astype('float64'), SWAN[17].astype('float64'))[0][1])
    MSE1.append(Root_Mean(MEAS[3],SWAN[3])[0])
    RMSE1.append(Root_Mean(MEAS[3],SWAN[3])[1])
    MSE2.append(Root_Mean(MEAS[4],SWAN[4])[0])
    RMSE2.append(Root_Mean(MEAS[4],SWAN[4])[1])
    MSE14.append(Root_Mean(MEAS[16],SWAN[16])[0])
    RMSE14.append(Root_Mean(MEAS[16],SWAN[16])[1])
    MSE15.append(Root_Mean(MEAS[17],SWAN[17])[0])
    RMSE15.append(Root_Mean(MEAS[17],SWAN[17])[1])
    STDEV1.append(statistics.stdev(SWAN[3]))
    STDEV2.append(statistics.stdev(SWAN[4]))
    STDEV14.append(statistics.stdev(SWAN[16]))
    STDEV15.append(statistics.stdev(SWAN[17]))
    STDEVm1.append(statistics.stdev(MEAS[3]))
    STDEVm2.append(statistics.stdev(MEAS[4]))
    STDEVm14.append(statistics.stdev(MEAS[16]))
    STDEVm15.append(statistics.stdev(MEAS[17]))
    bias1.append((np.sum(SWAN[3]-MEAS[3])/MEAS[3].size))
    bias2.append((np.sum(SWAN[4]-MEAS[4])/MEAS[4].size))
    bias14.append((np.sum(SWAN[16]-MEAS[16])/MEAS[16].size))
    bias15.append((np.sum(SWAN[17]-MEAS[17])/MEAS[17].size))
    var1.append(np.var([MEAS[3], SWAN[3]]))
    var2.append(np.var([MEAS[4], SWAN[4]]))
    var14.append(np.var([MEAS[16], SWAN[16]]))
    var15.append(np.var([MEAS[17], SWAN[17]]))
print(names_of_models)
print(Loc1_corr, Loc2_corr, Loc14_corr, Loc15_corr)
#print(MSE1,RMSE1,MSE2,RMSE2,MSE14,RMSE14,MSE15,RMSE15)
#print(STDEV1, STDEV2, STDEV14, STDEV15)
#print(STDEVm1, STDEVm2, STDEVm14, STDEVm15)
#print(bias1, bias2, bias14, bias15)
#print(var1, var2, var14, var15)
```
